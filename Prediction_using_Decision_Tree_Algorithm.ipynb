### Importing the dataset

```
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets

# Load the iris dataset
data = pd.read_csv("/Iris.csv")
data.head() # See the first 5 rows
```

```
data.shape  #Our dataset has 150 rows and 6 columns
```

### Preparing the data

```
X = data.iloc[:,[1,2,3,4]].values
y = data.iloc[:,5]
```

### Let's define Decision Tree and Visualize it

Since it's a classification problem we will use Decision Tree Classifier

```
# Defining the algorithm
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```
# Visualising the Decision Tree for understaning it better
# Importing the required libraries
!pip install pydotplus
!apt-get install graphviz -y;
print("Installation complete")
```

```
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus

# Visualise the graph
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,
                feature_names=data.iloc[:,[1,2,3,4]].columns,
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
